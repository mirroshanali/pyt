```python
import pandas as pd
import matplotlib.pyplot as plt
data = [['E1', 'M', 30, 100, 'Normal', 500],
['E2', 'F', 35, 75, 'Overweight', 350],
['E3', 'M', 25, 125, 'Underweight', 750],
['E4', 'F', 40, 50, 'Obesity', 400],
['E5', 'M', 20, 200, 'Fit', 1000]]
df = pd.DataFrame(data, columns = ['Emp_ID', 'Gender', 'Age', 'Sales','BMI', 'Income'])
df.hist()
plt.show()
```

```python
df.plot.bar()
plt.bar(df['Age'], df['Sales'])
plt.xlabel("Age")
plt.ylabel("Sales")
plt.show()
```

```python
df.plot.box()
plt.boxplot(df['Income'])
plt.show()
```

```python
plt.pie(df['Age'], labels = {"A", "B", "C", "D", "E"},
autopct ='% 1.1f %%', shadow = True)
plt.show()
```

```python
plt.scatter(df['Sales'], df['Age'])
plt.show()
```
